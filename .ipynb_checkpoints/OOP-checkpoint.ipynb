### Un clase denominada fracción 

```python
def mcd(m,n):
    while m%n != 0:
        mViejo = m
        nViejo = n

        m = nViejo
        n = mViejo%nViejo
    return n

class fraccion:
    def __init__(self, numerador, denominador):
        self.num = numerador
        self.den = denominador
        
    # Reescribiendo la función str para la clase    
    def __str__(self):
        # Imprime sin denominador, si este es 1
        if self.den == 1:
            return str(self.num)
        else:
            return str(self.num)+"/"+str(self.den)
    
    #Reescribiendo el operador +
    def __add__(self, otraFraccion):
            
            #Este es el criterio: f1.__add__(f2)
        nuevoNum = self.num*otraFraccion.den + self.den*otraFraccion.num
        nuevoDen = self.den * otraFraccion.den        
        comun = mcd(nuevoNum,nuevoDen)
        return fraccion(nuevoNum//comun,nuevoDen//comun)  # se utiliza el operador // porque genera un int      
#class fracción
    

f1 = fraccion(4,3)
f2 = fraccion(8,3)
f3 = f1 + f2
print("1 ***: {0}".format(f1))
print("2 ***: {0}".format(frac.__str__()))
print("3 ***: {0}".format(str(f1)))
print("4 ***: {0}".format(f3))






```
