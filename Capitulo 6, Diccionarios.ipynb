# Diccionarios

Son arreglos que permiten el acceso a información a través de **índice literal**, creado en el mismo elemento. 

### Creando el diccionario

```python
alien = {'color': 'verde', 'puntos': 5}
print(alien)
print("El indice es {0} y el contenido es {1}".format(alien['color'], alien['color']))
print("El indice es {0}\t y el contenido es {1}".format(alien['puntos'], alien['puntos']))
```

### Adicionando elementos
#### Se puede crear con cero elementos e ir adicionando: alien = {} y adicionar como está abajo

```python
alien['PosX'] = 760
alien['PosY'] = 480
print(alien)
```

### Recorriendo el diccionario

```python
user_0 = {'username': 'efermi', 'first': 'enrico','last': 'fermi'}

for key, value in user_0.items():
    print("Key:{0}, Valor:{1}".format(key, value))     
           
```

### Recorriendo con su clave de acceso o valor
#### Esta instrumento tiene .keys() y .values()

```python
gustos = {'marcelo': 'chocolate', 'magui': 'pizza', 'abner': 'pique macho', 'gustavo': 'hamburguesas', 'mariel': 'helados'}
clientes = ['marcelo', 'pedro', 'abner', 'mariel']
for nombre in gustos.keys():
    print(nombre.title())    
    if nombre in clientes: 
        print("Al cliente:{0} le gusta: {1}".format(nombre.title(), gustos[nombre]))
        
# Otra opcion podría ser
print("--------------------------------")
for cliente in clientes:
    if cliente in gustos.keys():
        print("Al cliente:{0} le gusta: {1}".format(cliente.title(), gustos[cliente]))

print("******************************")
for valor in gustos.values():
    print("{0}\t".format(valor)) # tambien sirve valor.title()

    
```

### Se puede usar el comando SET para eliminar los valores duplicados

```python
favorite_lenguajes = {'jen': 'python', 'sarah': 'c','edward': 'ruby','phil': 'python'}

for lenguaje in sorted(set(favorite_lenguajes.values())):  # adicioné SORTED para ver el resultado
    print(lenguaje.title())
    
```
