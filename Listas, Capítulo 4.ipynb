### Formatenado PRINT

```python
# combining positional and keyword arguments 
print('Number one portal is {0}, {1}, and {other}.'
     .format('Geeks', 'For', other ='Geeks')) 
  
# using format() method with number  
print("Geeks :{0:2d}, Portal :{1:8.2f}". 
      format(12, 00.546)) 
  
# Changing positional argument 
print("Second argument: {1:3d}, first one: {0:7.2f}". 
      format(47.42, 11)) 
  
print("Geeks: {a:5d},  Portal: {p:8.2f}". 
     format(a = 453, p = 59.058)) 
```

### Trabajo con listas

```python
motos = ['Yamaha','BMW','Suzuki', 'Honda']
autos = ['Toyota', 'Mercedes', 'Volkswagen', 'Peugot', 'Lada']
print(motos)
# Eliminando elementos de la LISTA
del motos[1]
print(motos)

# Uniendo dos listas
ma = motos + autos
print(ma)
print(ma[0][2])

#Trabajando como COLA
print(ma.pop())   # el último
print(ma.pop(0))  # el primero

#Removiendo elementos
print(ma)
ma.remove('Suzuki')
print(ma)

#Ordenando las LISTAS
print(autos)
print(sorted(autos))   #ordenamiento temporal
autos.sort()  # definitivo, también sirve autos.sort(reverse=True)
print(autos)
#invertir el orden de la lista
autos.reverse()
print(autos)
```

### Trabajo con índices 

```python

indices = list(range(1,12))
print(indices)

indices = list(range(1,30,4))
print(indices)

#Llenando la lista
cuadrados = [xx**2 for xx in range(1,10)]
print(cuadrados)

autos = ['Toyota', 'Mercedes', 'Volkswagen', 'Peugot', 'Lada']
print ("{0}: {1}".format(1, autos[0:3]))
print ("{0}: {1}".format(2, autos[:3]))
print ("{0}: {1}".format(3, autos[2:]))
```

### Copiando listas 

```python
otroAutos = autos
print(otroAutos)
```
