```python
import matplotlib.pyplot as plt
plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt


plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))                          #Roll genera valores de 0 a 4
performance = 3 + 10 * np.random.rand(len(people))          #Roll .rand = aleatorios corrientes ]0,1[
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center',
        color='green', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # esta de A a Z, ahora Z a A
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 30)
y = np.sin(4 * np.pi * x) * np.exp(-5 * x)

fig, ax = plt.subplots()

ax.fill(x, y, zorder=10, alpha=0.3)               #Roll llena el √°rea bajo, con transparencia 0.3

ax.plot(x,y)                                      #Roll une los puntos con l√≠nea
ax.scatter(x,y, marker="*", color="r"             #Roll coloca la figura
ax.grid(True, zorder=5)                           #Roll coloca el grid

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


color = 'cornflowerblue'
points = np.ones(5)  # Draw 5 points for each line
text_style = dict(horizontalalignment='right', verticalalignment='center',
                  fontsize=12, fontdict={'family': 'monospace'})


def format_axes(ax):
    ax.margins(0.2)
    #ax.set_axis_off()                            #roll permite visualizar los ejes


def nice_repr(text):
    return repr(text).lstrip('u')


# Plot all line styles.
fig, ax = plt.subplots()

linestyles = ['-', '--', '-.', ':']                 #roll definen el tipo de l√≠nea
for y, linestyle in enumerate(linestyles):
    ax.text(-0.1, y, nice_repr(linestyle), **text_style)
    ax.plot(y * points, linestyle=linestyle, color=color, linewidth=3)
    format_axes(ax)
    ax.set_title('line styles')

plt.show()
```
