```python
import numpy as np
import matplotlib.pyplot as plt

xLim = 360
yLim = 100
pasoX = 10
lc = [] # lista con valores constantes

fig1, ax1 = plt.subplots()

# llenando las listas **********************
for i in range(xLim):
    lc.append(yLim/2)
    
listSeno = np.arange(0, 360, pasoX)  
listaX = np.arange(0.0, xLim, 1)

# ajustes de la gráfica ********************
ax1.set_xlim(0, xLim*1.3)
ax1.set_ylim(0, yLim)
#ax1.set_aspect('equal')
ax1.set_title('Título de gráfica')
ax1.grid(True, color='b')
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Eje Y')

# colocando los datos en la gráfica *******************
plt.plot(listaX, -52+ listaX+listaX**2, color="r")
plt.plot(listaX, 23+ listaX, color="b")
plt.plot(listaX, lc, color="k")
plt.plot(listaX, yLim/2 + np.sin(listaX*np.pi/180)*yLim/3, color='g')
ax1.scatter(listSeno, yLim/2 + np.sin(listSeno*np.pi/180)*yLim/2, color='b')

# mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.tri as tri
import numpy as np

# First create the x and y coordinates of the points.
n_angles = 36
n_radii = 8
min_radius = 0.25
radii = np.linspace(min_radius, 0.95, n_radii)

angles = np.linspace(0, 2 * np.pi, n_angles, endpoint=False)
angles = np.repeat(angles[..., np.newaxis], n_radii, axis=1)
angles[:, 1::2] += np.pi / n_angles

x = (radii * np.cos(angles)).flatten()
y = (radii * np.sin(angles)).flatten()

# Create the Triangulation; no triangles so Delaunay triangulation created.
triang = tri.Triangulation(x, y)

# Mask off unwanted triangles.
triang.set_mask(np.hypot(x[triang.triangles].mean(axis=1),
                         y[triang.triangles].mean(axis=1))
                < min_radius)

fig1, ax1 = plt.subplots()
ax1.set_aspect('equal')
ax1.triplot(triang, 'bo-', lw=1)
ax1.set_title('triplot of Delaunay triangulation')
```

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out

# which you would then use as:

data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```
